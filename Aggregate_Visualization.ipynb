```python
# Code from Naive Bayesian
def distance_on_unit_sphere(lat1, long1, lat2, long2):

    # Convert latitude and longitude to 
    # spherical coordinates in radians.
    degrees_to_radians = math.pi/180.0
        
    # phi = 90 - latitude
    phi1 = (90.0 - lat1)*degrees_to_radians
    phi2 = (90.0 - lat2)*degrees_to_radians
        
    # theta = longitude
    theta1 = long1*degrees_to_radians
    theta2 = long2*degrees_to_radians
        
    # Compute spherical distance from spherical coordinates.
        
    # For two locations in spherical coordinates 
    # (1, theta, phi) and (1, theta, phi)
    # cosine( arc length ) = 
    #    sin phi sin phi' cos(theta-theta') + cos phi cos phi'
    # distance = rho * arc length
    
    cos = (math.sin(phi1)*math.sin(phi2)*math.cos(theta1 - theta2) + 
           math.cos(phi1)*math.cos(phi2))
    arc = math.acos( cos )

    # Remember to multiply arc by the radius of the earth 
    # in your favorite set of units to get length.
    return 3960*arc
```

```python
locs = zip(d['lat'], d['lng'])
dist = [distance_on_unit_sphere(42.3581,-71.0636,loc[0],loc[1]) for loc in locs]
d['dist_from_boston'] = dist
```

```python
def distance_variable_aggregate(y_col, x_col = 'dist_from_boston'):
    distances = d[x_col]
    feelings = d[y_col]
    interval_len = 300
    max_d = 3200
    num_steps = max_d/interval_len
    data = []
    for x in range(int(num_steps+1)):
        data.append([])
    for i, distance in enumerate(distances):
        if distance <= max_d:
            bucket = int(distance/interval_len)
            data[bucket].append(feelings[i])
    for i, sub_data in enumerate(data):
        sub_data = [value for value in sub_data if not math.isnan(value)]
        average = np.average(np.array(sub_data))
        data[i] = average
    x_range = range(0, int(max_d), interval_len)
    return (x_range, data)
```

```python
state2poly = defaultdict(list)

data = json.load(file("us-states.json"))
for f in data['features']:
    state = states_abbrev[f['id']]
    geo = f['geometry']
    if geo['type'] == 'Polygon':
        for coords in geo['coordinates']:
            state2poly[state].append(coords)
    elif geo['type'] == 'MultiPolygon':
        for polygon in geo['coordinates']:
            state2poly[state].extend(polygon)

def draw_state(plot, stateid, **kwargs):
    """
    draw_state(plot, stateid, color=..., **kwargs)
    
    Automatically draws a filled shape representing the state in
    subplot.
    The color keyword argument specifies the fill color.  It accepts keyword
    arguments that plot() accepts
    """
    for polygon in state2poly[stateid]:
        xs, ys = zip(*polygon)
        plot.fill(xs, ys, **kwargs)

        
def make_map(states, label, leftover_states = []):
    """
    Draw a cloropleth map, that maps data onto the United States
    
    Inputs
    -------
    states : Column of a DataFrame
        The value for each state, to display on a map
    label : str
        Label of the color bar

    Returns
    --------
    The map
    """
    fig = plt.figure(figsize=(12, 9))
    ax = plt.gca()

    cmap = cm.binary
    vmin, vmax = states.min(), states.max()
    norm = mpl.colors.Normalize(vmin=vmin, vmax=vmax)
    
    skip = set(['National', 'Hawaii', 'Alaska', 'District of Columbia', 'Guam', 'Puerto Rico',
                'Virgin Islands', 'American Samoa', 'Northern Mariana Islands'] + leftover_states)
    for state in states_abbrev.values():
        if state in skip:
            continue
        color = cmap(norm(states.ix[state]))
        draw_state(ax, state, color = color, ec='k')

    #add an inset colorbar
    ax1 = fig.add_axes([0.45, 0.70, 0.4, 0.02])    
    cb1=mpl.colorbar.ColorbarBase(ax1, cmap=cmap,
                                  norm=norm,
                                  orientation='horizontal')
    ax1.set_title(label)
    remove_border(ax, left=False, bottom=False)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlim(-180, -60)
    ax.set_ylim(15, 75)
    return ax
```

```python
df = pd.read_csv("GEO.csv")

def plot_metric(metric):
    metric_state_df = df[[metric, 'geo_admin1']].groupby('geo_admin1').groups
    metric_scores = df[metric]
    states = metric_state_df.keys()
    metric_score_by_state = {}
    for state in states:
        indices_for_state = metric_state_df[state]
        state_scores = []
        for i in indices_for_state:
            indiv_score = metric_scores[i]
            if not (math.isnan(indiv_score)):
                state_scores.append(indiv_score)
        metric_score_by_state[state] = np.average(state_scores)
    metric_df = pd.DataFrame.from_dict(metric_score_by_state, orient='index')
    metric_df.columns = ["metric_name"]
    make_map(metric_df.metric_name, metric)

for col in  cols_of_interest:
    plot_metric(col)
```

```python
""" Map for emo values for each day"""

all_locs_by_date = by_day.geo_admin1.values
dates = all_locs_by_date.keys()
"""change .emo.values to see other parameter"""
for day_num, data in enumerate(by_day.emo.values):
    indiv_date = dates[day_num]
    locations_for_day_num = all_locs_by_date[indiv_date]
    indiv_refined_locs = []
    indiv_refined_data = []
    # removes nans and irrelevant data
    for i, x in enumerate(data):
        if not math.isnan(x):
            indiv_refined_data.append(x)
            indiv_loc = locations_for_day_num[i]
            indiv_refined_locs.append(indiv_loc)
    # makes list of values by state per day
    state_by_day = {}
    for i, state in enumerate(indiv_refined_locs):
        assoc_data = indiv_refined_data[i]
        if state_by_day.has_key(state):
            state_by_day[state].append(assoc_data)
        else:
            state_by_day[state] = [assoc_data]
   
    # add leftover states for completeness
    state_names_for_day = state_by_day.keys()
    all_states = set(states_abbrev.values())
    leftover_states_for_day = list(all_states - set(state_names_for_day))
    # aggregate and create map
    for state in state_names_for_day:
        state_by_day[state] = np.average(state_by_day[state])
    metric_df_day = pd.DataFrame.from_dict(state_by_day, orient='index')
    metric_df_day.columns = ["metric_name"]
    metric = str("emo") + str(day_num)
    make_map(metric_df_day.metric_name, metric, leftover_states_for_day)
            
```
