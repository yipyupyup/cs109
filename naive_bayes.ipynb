```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import dateutil
%matplotlib inline
pd.set_option('display.width',500)
pd.set_option('display.max_columns',100)
from scipy.stats.stats import pearsonr
import os
import string
from os import listdir
from os.path import isfile, join
import nltk
#nltk.download() # enter d, then wordnet
import nltk.tokenize as tok
from nltk.corpus import wordnet as wn
from nltk.corpus import stopwords as stop
import readabilitytests as rt
import math
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cross_validation import train_test_split
from sklearn.naive_bayes import MultinomialNB
import dateutil
```

```python
d = pd.read_csv("geo_tweets_content.csv", parse_dates=[0], index_col=0)
d = d.drop(d.index[[3064, 3063,3062]])
d = d.drop(d.index[[6167, 6671, 10842, 15427]])
d = d.drop(d.index[[9269,14391,27207]])
d = d[d['geo_admin0'] == 'United States']
d = d[(d['geo_admin1'] != "Puerto Rico") & (d['geo_admin1'] != 'Alaska') & (d['geo_admin1'] != 'Hawaii')]
d = d.dropna(axis=0, how='any', subset=['lat', 'lng', 'geo_admin1'])
d = d.convert_objects(convert_numeric=True)

# For some reason pandas isn't parsing the dates correctly
# on its own so I do it manually here
dates = [dateutil.parser.parse(str(x)) for x in d.index]
d['date'] = dates
d.set_index('date', inplace=True)
d.index.max()
```

```python
def distance_on_unit_sphere(lat1, long1, lat2, long2):

    # Convert latitude and longitude to 
    # spherical coordinates in radians.
    degrees_to_radians = math.pi/180.0
        
    # phi = 90 - latitude
    phi1 = (90.0 - lat1)*degrees_to_radians
    phi2 = (90.0 - lat2)*degrees_to_radians
        
    # theta = longitude
    theta1 = long1*degrees_to_radians
    theta2 = long2*degrees_to_radians
        
    # Compute spherical distance from spherical coordinates.
        
    # For two locations in spherical coordinates 
    # (1, theta, phi) and (1, theta, phi)
    # cosine( arc length ) = 
    #    sin phi sin phi' cos(theta-theta') + cos phi cos phi'
    # distance = rho * arc length
    
    cos = (math.sin(phi1)*math.sin(phi2)*math.cos(theta1 - theta2) + 
           math.cos(phi1)*math.cos(phi2))
    arc = math.acos( cos )

    # Remember to multiply arc by the radius of the earth 
    # in your favorite set of units to get length.
    return 3960*arc
```

```python
locs = zip(d['lat'], d['lng'])
dist = [distance_on_unit_sphere(42.3581,-71.0636,loc[0],loc[1]) for loc in locs]
d['dist_from_boston'] = dist
```

```python
def make_xy(d, vectorizer=None):
    text = list(d.text)    
    if vectorizer is None:
        vectorizer = CountVectorizer(min_df=0)
    vectorizer.fit(text)
    x = vectorizer.transform(text)
    y = d.dist_from_boston
    return x.tocsc(),np.array(y)
```

```python
X, Y = make_xy(d)
clf = MultinomialNB()
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, train_size=0.3)
clf.fit(xtrain, ytrain)
xtrain2, xtest2, ytrain2, ytest2 = train_test_split(xtest, ytest, train_size=0.5)
accuracy_test = clf.score(xtest2,ytest2)*100 #### maybe not the right measure of accuracy
print 'Accuracy of the testing set:\t%2.2f%%' %accuracy_test
```
