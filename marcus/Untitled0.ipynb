```python
import requests
import json
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
from pprint import pprint
import datetime
import numpy as np
import us
import csv
import re
from collections import defaultdict
```

```python
# The API only returns 500 tweets at a time, but you can look at small slices
# of time to get zillions of tweets

# And the times need to be in UTC format
def utc_from_dt(dt):
    return int((dt - datetime.datetime(1970, 1, 1)).total_seconds())

def dt_from_utc(utc):
    return datetime.datetime(1970, 1, 1) + datetime.timedelta(seconds=utc)
```

```python
with open("by_city.json", "rb") as f:
    city_data = json.load(f)
with open("by_state.json", "rb") as f:
    state_data = json.load(f)
```

```python
tweet_list = []
for state in city_data:
    for city in city_data[state]:
        for tweet in city_data[state][city]:
            tweet_list.append(tweet)
            
```

```python
cf = pd.DataFrame.from_records(tweet_list)
```

```python
cf = cf.dropna()
cf = cf.convert_objects(convert_numeric=True)
```

```python
by_county = cf.groupby(['date','fips_code'])
```

```python
print by_county.head()
```

```python
aggy = by_county.agg({'baseline':np.sum, 'mentions':np.sum})
```

```python
aggy['frac'] = np.array(aggy['mentions'], dtype=float)/aggy['baseline']
```

```python
aggy.head()
```

```python
what_i_want = aggy['frac'].to_dict()
```

```python
what_i_want.to_dict()
```

```python
x = defaultdict(dict)
for key, value in what_i_want.iteritems():
    x[key[0]][key[1]] = value
```
